```python
RULES = [
    lambda df: df['VariantEffect'] != 'synonymous',
    lambda df: df['VariantLocation'] != 'intron',
    lambda df: np.invert(df['1000g_AF'] > 0.01),
    lambda df: np.invert(df['gnomadAltFreq_all'] > 0.01)
]


def rules_filter(df):
    for rule in RULES:
        df = df[rule(df)]
    return df
```
